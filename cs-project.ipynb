### 3. For each (primary) breed, determine the ratio between the number of dogs of Mixed Breed and those not of Mixed Breed

```python
breed_secondary   = df['breed_secondary'].value_counts()




df_grouped = df.groupby(['breed_primary', 'breed_secondary']).size().reset_index(name='count')


# Define the year to filter
breed_to_filter = 'Mixed Breed'

# Filter rows based on the year
df_Mixed = df_grouped[df_grouped['breed_secondary'] == breed_to_filter]

# Sort the filtered DataFrame by count in descending order
df_Mixed_sorted = df_Mixed.sort_values(by='count', ascending=False)

print(f"Number of Mixed Breed Dogs is: {df_Mixed_sorted['count'].sum()}")



df_Not_Mixed = df_grouped[df_grouped['breed_secondary'] != breed_to_filter]
df_Not_Mixed_sorted = df_Not_Mixed.sort_values(by='count', ascending=False)
# print(df_sum[df_sum['breed_secondary'] == breed_to_filter]['total_count'].values[0])

# Print the sorted filtered DataFrame
print(f"Number of Not Mixed Breed Dogs hs: {df_Not_Mixed_sorted['count'].sum()}")

print(df_Mixed_sorted)
```

### 4. For each (primary) breed, determine the earliest and the latest posted timestamp.

### 5. For each state, compute the sex imbalance, that is the difference between male and female dogs. In which state this imbalance is largest?

```python
counts = df.groupby(['contact_state', 'sex']).size().reset_index(name='count')
pivoted = counts.pivot(index='contact_state', columns='sex', values='count')
pivoted['imbalance'] = abs(pivoted['Male'] - pivoted['Female'])

# print the pivoted DataFrame
print(pivoted.sort_values(by='imbalance', ascending=False))
```

### 7. Find the dogs involved in at least 3 travels. Also list the breed of those dogs.
<br>
<br>
The dataset consist of each travel for dogs. each dog has a unigue ID , so we can count the number of travels based on the ID.<br>
after detemining the number of travels for each dog in this dataset, we can join the results from this dataset with the original dataset based on the ID.

```python
df2 = pd.read_csv('dataset/dogTravel.csv')

id_counts = df2['id'].value_counts()

# Filter the result to only include id of dogs with at least 3 travels
id_counts_filtered = id_counts[id_counts > 2]


# Create a new DataFrame from the result of value_counts() and rename the columns
id_counts_df = pd.DataFrame({'id': id_counts_filtered.index, 'travels': id_counts_filtered.values})

# print(id_counts_df)


# Join the filtered dogs with the breed dataset on the 'dog_id' column
filtered_dogs_breed = pd.merge(id_counts_df, df, on='id')

# print(filtered_dogs_breed)

result_df =filtered_dogs_breed[['id', 'travels','breed_primary' ]]

# Print the result
print(result_df)
```

### 8. Fix the travels table so that the correct state is computed from the manual and the found fields. If manual is not missing, then it overrides what is stored in found.

```python
import numpy as np


df2['correct state'] = np.where(df2['manual'].notnull(), df2['manual'], df2['found'])

# Print the result
print(df2[['manual','found','correct state']]).head(5)
```

### 9. For each state, compute the ratio between the number of travels and the population.


```python
df3 = pd.read_csv('dataset/NST-EST2021-POP.csv')

state_counts = df2['contact_state'].value_counts()

state_travel_df = pd.DataFrame({'contact_state': state_counts.index, 'travels': state_counts.values})

state_travel_population = pd.merge(state_travel_df, df3, on='contact_state')

state_travel_population['population'] = state_travel_population['population'].str.replace('.', '',regex=True).astype(float)
state_travel_population['travels'] = state_travel_population['travels'].astype(float)

state_travel_population['ratio10e6']= round((state_travel_population['travels']/state_travel_population['population'])*1000000)


state_travel_population.sort_values(by='ratio10e6', ascending=False)
```
